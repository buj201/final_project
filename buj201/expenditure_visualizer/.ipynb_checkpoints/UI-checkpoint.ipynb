## Introduction

### Data Sources
This tool allows the user to interactively explore NYC school expenditures from 2006-2012. Data sources include:
1. The <a href='https://data.cityofnewyork.us/Education/School-Demographics-and-Accountability-Snapshot-20/ihfw-zy9j'> NYC School Demographics and Accountability Snapshot 2006-2012 </a>, which includes demographic information for all NYC Public schools from 2006-2012;
2. The Research Alliance for New York City Schools <a href='http://steinhardt.nyu.edu/research_alliance/research/schma'>School-Level Master File (SCHMA)</a>, which (in this work) is only used to obtain geographic coordinates for schools from 2010-2012;
3. NYC <a href='http://schools.nyc.gov/Offices/DBOR/SBER/default.htm'>School-Based Expenditure Reports</a>, which were scraped to get the expenditure data used for all visualizations.

### Uses
The purpose of this tool is to allow users to visually explore:
1. Differences in funding across schools, and the relationship between funding and school characteristics (described <a href='http://schools.nyc.gov/Offices/DBOR/SBER/costvary.htm'> here by NYC Department of Education</a>).
2. Expenditures, by category, for a specific school across the years 2006-2012. For context, the majority of schools' budgets are discretionary, with funds used at the principal's discretion (see chart below, from <a href='http://schools.nyc.gov/offices/d_chanc_oper/budget/dbor/allocationmemo/fy12_13/FY13_PDF/FSF_Guide.pdf'>Fair Student Funding and School Budget Resource Guide </a>).

## Visualization tools
### PCA Explorer

In this tool, Principal Components Analysis (PCA) is used to reduce the dimensionality of the expenditure data from approximately 50 features to 2. These two features don't correspond to specific line-items in the expenditure report- intead, they (loosely) show the two directions where schools have the greatest variation in the expenditures. The tool is useful since it can be used to explore the expenditure data and see if there are (visually apparent) trends or clusters based on school characteristics (focusing on the demographic characteristics in the School Demographic and Accountability Snapshot).

```python
%matplotlib inline
run 'pca_to_explore_budget_data.py'
```

### School Dashboard

Using this tool, you can search by a school name to see how that school's budget was distributed across different expenditure categories (i.e. Building Services vs. Classroom Instruction). In addition to the input school, a comparison group is shown in all plots. This comparison group is defined for a given year:
- If, in that year, the input school was Title-1, the comparison group is the average of all Title-1 NYC schools.
- If, in that year, the input school was not Title-1, the comparison group is the average of all non-Title-1 NYC schools.

```python
%run 'school_dashboard.py'
```

### School Expenditure Map

Using this tool, you can map all NYC schools (for an input year between 2010 and 2012- years where geographic coordinants are available for schools in the SCHMA), coloring by one of the major categories in the expenditure report. Using this map, you can see how the the percent of school expenditures on a given category vary by borough and neighborhood.

Note the map initially takes several minutes to load, as it downloads a (fairly high) resolution basemap of the NYC area.
