## Introduction

### Data Sources
This tool allows the user to interactively explore NYC school expenditures from 2006-2012. Data sources include:
1. The <a href='https://data.cityofnewyork.us/Education/School-Demographics-and-Accountability-Snapshot-20/ihfw-zy9j'> NYC School Demographics and Accountability Snapshot 2006-2012 </a>, which includes demographic information for all NYC Public schools from 2006-2012;
2. The Research Alliance for New York City Schools <a href='http://steinhardt.nyu.edu/research_alliance/research/schma'>School-Level Master File (SCHMA)</a>, which (in this program) is only used to obtain geographic coordinates for schools from 2010-2012;
3. NYC <a href='http://schools.nyc.gov/Offices/DBOR/SBER/default.htm'>School-Based Expenditure Reports</a>, which were scraped to get the expenditure data used for all visualizations.

### Uses
The purpose of this tool is to allow users to visually explore:
1. Differences in funding across schools, and the relationship between funding and school characteristics (for more explanation on the origin of these differences, see <a href='http://schools.nyc.gov/Offices/DBOR/SBER/costvary.htm'> this explantion from the NYC Department of Education</a>).
2. Expenditures, by category, for a specific school across the years 2006-2012. For context, the majority of schools' budgets are discretionary, with funds used at the principal's discretion (see chart below, from <a href='http://schools.nyc.gov/offices/d_chanc_oper/budget/dbor/allocationmemo/fy12_13/FY13_PDF/FSF_Guide.pdf'>Fair Student Funding and School Budget Resource Guide </a>).

## Visualization tools
### PCA Explorer

In the first tool, Principal Components Analysis (PCA) is used to reduce the dimensionality of the expenditure data from approximately 50 features to just 2. These two features don't correspond to specific line-items in a school's expenditure report. Instead, (speaking rather informally) they are the two dimensions in which school expenditures have the greatest variation (so schools are spread apart as much as possible in just two dimensions). This allows the user to more readily see if there are (visually apparent) trends or clusters based on school characteristics.

### School Expenditure Boxplots

While principal components analysis allows us to visualize school expenditures in a way that reveals the maximum spread of the school expenditure data (in two dimensions), it does not allow us to make comparison between groups of schools on specific expenditure types. Fortunately, this exploration is possible using the tool below. To use this tool:
- Select an expenditure category.
- Select a demographic or school characteristic by which to group the data.
- Check the checkbox to plot the expenditure as a percent of a school's total expenditures for a given year, instead of as a dollar amount.

For reference, note the available list of expenditure categories is a subset of all expenditures listed in the NYC School Expenditure reports. It includes what are perhaps the most intuitively understood expenditure categories (for example, excluding expenditures related to central district administration).

```python
%run 'box_plot.py'
```

### School Dashboard

The two preceeding tools allow you to explore groups of schools. In contrast, this tool allows you search by a school name to see how that school's budget was distributed across different expenditure categories (i.e. Building Services vs. Classroom Instruction). In addition to the input school, a comparison group is shown in all plots (in green in the top plot, as hatched bars in the bottom bar plots). For a given year, this comparison group is determined by the searched school's title 1 status:
- If the input school was Title-1 in that year, the comparison group is the average of all Title-1 NYC schools.
- If the input school was not Title-1 in that year, the comparison group is the average of all non-Title-1 NYC schools.
